```python
import pandas as pd
import requests
import json
```

```python
fpath = 'roster_and_stats_merged.csv'
stats_df = pd.read_csv(fpath,index_col=0)
```

```python
stats_df['Player_url'] = stats_df['Player Name'].replace(' ','%25',regex=True)
stats_df.head()
```
